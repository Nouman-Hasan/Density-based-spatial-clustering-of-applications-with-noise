```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
import plotly.express as px
```

```python
sns.set_theme(color_codes=True)
sns.set_style('whitegrid')
%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
circles = pd.read_csv('cluster-circles.csv')
```

```python
px.scatter(circles,x = 'X1' ,y = 'X2' , height=500)
```

```python
def display_categories(model,data):
    
    labels =  model.fit_predict(data)
    fig =  px.scatter(data , x = 'X1' , y = 'X2' , color=labels, height=500)
    fig.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
display_categories(model,blobs)
```

```python
model = KMeans(n_clusters=2)
```

```python
display_categories(model,moons)
```

```python
model = KMeans(n_clusters=2)
```

```python
display_categories(model,circles)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN()
```

```python
#help(DBSCAN)
```

```python
display_categories(model,blobs)
```

```python
model = DBSCAN(eps=0.2)
```

```python
display_categories(model,moons)
```

```python
display_categories(model,circles)
```
